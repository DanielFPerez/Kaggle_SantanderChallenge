```python
import include.preprocess as preprocess
import include.printAssistant as printAssistant
```

```python
# Load a pre-cleaned the dataset
df_raw = preprocess.get_raw_data()
df_raw.head()
```

```python
df_filtered = preprocess.filter_datframe(dataframe=df_raw, filter_case="high-valued")
df_filtered.head()
# df_filtered.shape
```

```python
df_normalized = preprocess.norm_transform_dataframe(dataframe=df_filtered, norm_case="minmax", scale_factor=1000)
df_normalized.head()
```

```python
# Apply PCA 
df_pca = preprocess.apply_pca(dataframe=df_normalized, dimensions=15)
df_pca.head()
```

```python
# Split data
X_train, X_test, y_train, y_test = preprocess.split_datagram_test_train(df_pca, test_size=0.1)
print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test shape: ", y_test.shape)
print("Number of 1-classes in y_train before smote: ", y_train[y_train == 1].shape)
```

```python
# Perform oversampling
X_smot, y_smot = preprocess.perform_smote(X_train, y_train)
print("X_smot shape: ", X_smot.shape)
print("y_smot shape: ", y_smot.shape)
print("Number of 1-classes in y_smot: ", y_smot[y_smot == 1].shape)
```

```python

from xgboost import XGBClassifier

# For evaluating a model
from sklearn.metrics import auc, roc_curve

model = XGBClassifier()
model.fit(X_smot, y_smot)
y_pred = model.predict(X_test)

fpr, tpr, thresholds = roc_curve(y_test, y_pred)

roc_auc = auc(fpr,tpr)
print("AUC value is: ", roc_auc)

printAssistant.plot_rco_auc(fpr, tpr, roc_auc)
```

```python
from scipy import interp
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import StratifiedKFold
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

cv = StratifiedKFold(n_splits=4)
classifier = XGBClassifier()

tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)

i = 0
for train, test in cv.split(X_smot, y_smot):
    print("Train: ",train)
    print("Test: ", test)
    predicted = classifier.fit(X_smot[train], y_smot[train]).predict(X_smot[test])
    # Compute ROC curve and area the curve
    fpr, tpr, thresholds = roc_curve(y_smot[test], predicted)
    tprs.append(interp(mean_fpr, fpr, tpr))
    tprs[-1][0] = 0.0
    roc_auc = auc(fpr, tpr)
    aucs.append(roc_auc)
    plt.plot(fpr, tpr, lw=1, alpha=0.3,
             label='ROC fold %d (AUC = %0.2f)' % (i, roc_auc))

    i += 1
plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
         label='Chance', alpha=.8)

mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs)
plt.plot(mean_fpr, mean_tpr, color='b',
         label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
         lw=2, alpha=.8)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
plt.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2,
                 label=r'$\pm$ 1 std. dev.')

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```
